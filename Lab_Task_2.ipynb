```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
def RGB_to_Gray(image):
  r, c, ch = image.shape

  abc = np.zeros((r, c), dtype=np.uint8)
  for i in range(r):
    for j in range(c):
      r, g, b = image[i, j]
      abc[i, j] = r*0.29 + g*0.58 + b*0.11

  return abc

img2 = cv2.imread('/content/sample_data/DIP_Lab_task_02.jpg')
gray_img = RGB_to_Gray(img2)

plt.figure(figsize=(5, 5))
plt.subplot(1, 2, 1)
plt.imshow(img2, cmap='gray')
plt.axis('off')
plt.title('Given image:')
plt.subplot(1, 2, 2)
plt.imshow(gray_img, cmap='gray')
plt.axis('off')
plt.title('Updated Image:')
```
