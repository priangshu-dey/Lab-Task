```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

def Gamma_Trans(Pix_Data, Gamma, Max_Clr):
    c = 1  # Scaling constant, typically 1
    # Apply gamma transformation to each pixel (RGB channels)
    for i in range(0, len(Pix_Data), 3):
        for j in range(3):
            # Apply the transformation and clamp to max_color range
            Pix_Data[i + j] = int(c * (Pix_Data[i + j] / Max_Clr) ** Gamma * Max_Clr)
    return Pix_Data

img = cv2.imread('/content/sample_data/DIP_Lab_task_04.jpg')


Drk_img = Gamma_Trans(img.flatten(), 5, 255)
Drk_img = Drk_img.reshape(img.shape)
cv2_imshow(img)
cv2_imshow(Drk_img)
```
