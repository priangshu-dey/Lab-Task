```python
import cv2
import matplotlib.pyplot as plt
import numpy as np



#image smoothing
def conv(image, kernel):
  kernel_Height, kernel_Width = kernel.shape
  pad_Height = kernel_Height//2
  pad_Width = kernel_Width//2

  padded_img = np.pad(image, ((pad_Height, pad_Height), (pad_Width, pad_Width)), 'constant')
#It is needed to take 0 padding for filtering
  r, c = image.shape
  output = np.zeros((r, c))

  for i in range(r):
    for j in range(c):
        roi = padded_img[i:i+kernel_Height, j:j+kernel_Width]
        output[i, j] = np.median(roi)

  return output

img_3 = cv2.imread('/content/sample_data/DIP_Lab_task_03.png')

plt.imshow(img_3)
plt.axis('off')
plt.title('original Image:')
plt.show()
#Salt and Paper noise
img_3_gray = cv2.cvtColor(img_3, cv2.COLOR_BGR2GRAY)
smoothing_kernel = np.ones((5, 5), dtype=np.uint8)

Updated_img = conv(img_3_gray, smoothing_kernel)


plt.imshow(Updated_img, cmap = 'gray')
plt.axis('off')
plt.title('Updated Image:')
plt.show()
```
